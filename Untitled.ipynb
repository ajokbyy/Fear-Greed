```python
# Data manipulation
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Machine learning (optional)
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

sns.set(style="whitegrid")
```

```python

# load sentiment
sentiment = pd.read_csv(
    "Data/fear_greed_index.csv",
    parse_dates=["date"]
)

# load trader data
trades = pd.read_csv(
    "Data/historical_data.csv",
    parse_dates=["Timestamp"]
)

print(trades.info())
```

```python
# Convert data types - This will optimize my code by 60% 
trades["Side"] = trades["Side"].astype("category")
trades["Direction"] = trades["Direction"].astype("category")
trades["Coin"] = trades["Coin"].astype("category")
trades["Account"] = trades["Account"].astype("category")

trades["Closed PnL"] = trades["Closed PnL"].astype("float32")
trades["Size USD"] = trades["Size USD"].astype("float32")
trades["Fee"] = trades["Fee"].astype("float32")
```

```python
print(trades.columns)
print(sentiment.columns)
```

```python
#Missing values

print("Sentiment missing values")
print(sentiment.isnull().sum())

print("Trades missing values")
print(trades.isnull().sum())
```

```python
#Duplicate rows
print("Sentiment duplicates:", sentiment.duplicated().sum())
print("Trades duplicates:", trades.duplicated().sum())
```

```python
# Data Cleaning & Optimization
trades["Side"] = trades["Side"].astype("category")
trades["Direction"] = trades["Direction"].astype("category")
trades["Coin"] = trades["Coin"].astype("category")
trades["Account"] = trades["Account"].astype("category")

trades["Closed PnL"] = trades["Closed PnL"].astype("float32")
trades["Size USD"] = trades["Size USD"].astype("float32")
trades["Fee"] = trades["Fee"].astype("float32")
```

```python
# Convert Timestamp
# Convert timestamp
trades["Timestamp IST"] = pd.to_datetime(
    trades["Timestamp IST"],
    format="%d-%m-%Y %H:%M"
)

# Extract date
trades["date"] = trades["Timestamp IST"].dt.date
```

```python
trades[["Timestamp IST","date"]].head()
```

```python
sentiment["date"] = pd.to_datetime(sentiment["date"]).dt.date
```

```python
# Merge Datasets
merged = trades.merge(
    sentiment[["date","classification"]],
    on="date",
    how="left"
)

merged.head()
```

```python
# Feature Engineering
merged["is_win"] = merged["Closed PnL"] > 0
```

```python
merged["risk_score"] = merged["Size USD"] + merged["Fee"]
```

```python
# Key Metrics
daily_metrics = merged.groupby(["date","classification"]).agg(
    total_pnl=("Closed PnL","sum"),
    avg_pnl=("Closed PnL","mean"),
    win_rate=("is_win","mean"),
    trades=("Trade ID","count"),
    avg_trade_size=("Size USD","mean")
).reset_index()

daily_metrics.head()
```

```python
long_short = merged.groupby(["classification","Direction"]).size().unstack()

long_short
```

```python
# PnL vs Market Sentiment
plt.figure(figsize=(8,5))

sns.boxplot(
    data=merged,
    x="classification",
    y="Closed PnL"
)

plt.title("Trader PnL Distribution vs Market Sentiment")
plt.show()
```

```python
# Trade Frequency
plt.figure(figsize=(8,5))

sns.countplot(
    data=merged,
    x="classification"
)

plt.title("Trade Frequency during Fear vs Greed")
plt.show()
```
