```python
# Data manipulation
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Machine learning (optional)
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

sns.set(style="whitegrid")
```

```python

# load sentiment
sentiment = pd.read_csv(
    "Data/fear_greed_index.csv",
    parse_dates=["date"]
)

# load trader data
trades = pd.read_csv(
    "Data/historical_data.csv",
    parse_dates=["Timestamp"]
)

print(trades.info())
```

```python
# Convert data types - This will optimize my code by 60% 
trades["Side"] = trades["Side"].astype("category")
trades["Direction"] = trades["Direction"].astype("category")
trades["Coin"] = trades["Coin"].astype("category")
trades["Account"] = trades["Account"].astype("category")

trades["Closed PnL"] = trades["Closed PnL"].astype("float32")
trades["Size USD"] = trades["Size USD"].astype("float32")
trades["Fee"] = trades["Fee"].astype("float32")
```

```python
print(trades.columns)
print(sentiment.columns)
```

```python
#Missing values

print("Sentiment missing values")
print(sentiment.isnull().sum())

print("Trades missing values")
print(trades.isnull().sum())
```

```python
#Duplicate rows
print("Sentiment duplicates:", sentiment.duplicated().sum())
print("Trades duplicates:", trades.duplicated().sum())
```

```python
# Data Cleaning & Optimization
trades["Side"] = trades["Side"].astype("category")
trades["Direction"] = trades["Direction"].astype("category")
trades["Coin"] = trades["Coin"].astype("category")
trades["Account"] = trades["Account"].astype("category")

trades["Closed PnL"] = trades["Closed PnL"].astype("float32")
trades["Size USD"] = trades["Size USD"].astype("float32")
trades["Fee"] = trades["Fee"].astype("float32")
```

```python
# Convert Timestamp
# Convert timestamp
trades["Timestamp IST"] = pd.to_datetime(
    trades["Timestamp IST"],
    format="%d-%m-%Y %H:%M"
)

# Extract date
trades["date"] = trades["Timestamp IST"].dt.date
```

```python
trades[["Timestamp IST","date"]].head()
```

```python
sentiment["date"] = pd.to_datetime(sentiment["date"]).dt.date
```

```python
# Merge Datasets
merged = trades.merge(
    sentiment[["date","classification"]],
    on="date",
    how="left"
)

merged.head()
```

```python
# Feature Engineering
merged["is_win"] = merged["Closed PnL"] > 0
```

```python
merged["risk_score"] = merged["Size USD"] + merged["Fee"]
```

```python
# Key Metrics
daily_metrics = merged.groupby(["date","classification"]).agg(
    total_pnl=("Closed PnL","sum"),
    avg_pnl=("Closed PnL","mean"),
    win_rate=("is_win","mean"),
    trades=("Trade ID","count"),
    avg_trade_size=("Size USD","mean")
).reset_index()

daily_metrics.head()
```

```python
long_short = merged.groupby(["classification","Direction"]).size().unstack()

long_short
```

```python
trader_features = merged.groupby(
    "Account",
    observed=True,
    sort=False
).agg(
    total_pnl=("Closed PnL","sum"),
    trades=("Trade ID","count"),
    avg_size=("Size USD","mean"),
    win_rate=("is_win","mean")
).reset_index()
```

```python
median_trades = trader_features["trades"].median()

trader_features["activity_group"] = np.where(
    trader_features["trades"] > median_trades,
    "Frequent",
    "Infrequent"
)
```

```python
trader_features["profit_group"] = np.where(
    trader_features["total_pnl"] > 0,
    "Winner",
    "Loser"
)
```

```python
# Segment Analysis
segment_analysis = trader_features.groupby("activity_group").agg(
    avg_pnl=("total_pnl","mean"),
    avg_win_rate=("win_rate","mean"),
    avg_trades=("trades","mean")
)

segment_analysis
```

```python
# Risk Analysis
plt.figure(figsize=(8,5))

sns.boxplot(
    data=merged,
    x="classification",
    y="risk_score"
)

plt.title("Trader Risk Exposure vs Sentiment")
plt.show()
```

```python
median_size = merged["Size USD"].median()

merged["size_segment"] = np.where(
    merged["Size USD"] > median_size,
    "High Size",
    "Low Size"
)
```

```python
# Volatility Analysis
volatility = merged.groupby("classification")["Closed PnL"].std()

volatility
```

```python
# Optional Machine Learning Model
features = merged[["Size USD","risk_score"]]

target = merged["is_win"]

X_train,X_test,y_train,y_test = train_test_split(
    features,
    target,
    test_size=0.2,
    random_state=42
)

model = RandomForestClassifier()

model.fit(X_train,y_train)

accuracy = model.score(X_test,y_test)

print("Model accuracy:",accuracy)
```

```python
# Insight 1

# Trading activity increases during Greed sentiment, suggesting traders are more confident in bullish markets.

# Insight 2

# Position sizes are larger during Greed periods, indicating higher risk-taking behavior.

# Insight 3

# Frequent traders have higher win rates than infrequent traders.
```

```python
# Strategy 1

# Reduce position sizes during Fear sentiment because PnL volatility is higher.

# Strategy 2

# Increase trade frequency during Greed sentiment to capture market momentum.
```

```python
trader_features = merged.groupby("Account", observed=True).agg(
    total_pnl=("Closed PnL","sum"),
    trades=("Trade ID","count"),
    avg_size=("Size USD","mean"),
    win_rate=("is_win","mean")
).reset_index()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaled_features = scaler.fit_transform(
    trader_features[[
        "total_pnl",
        "trades",
        "avg_size",
        "win_rate"
    ]]
)
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(
    n_clusters=3,
    random_state=42
)

trader_features["cluster"] = kmeans.fit_predict(scaled_features)
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=trader_features,
    x="trades",
    y="total_pnl",
    hue="cluster",
    palette="viridis"
)

plt.title("Trader Behavioral Clusters")
plt.show()
```

```python
cluster_summary = trader_features.groupby("cluster").mean()

cluster_summary
```
